## Combined Color & Gradient Threshold for Lane Line Detection

```python
# Import Dependencies
import cv2
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.image import imread
%matplotlib inline
```

```python
# Read Image
img = imread('./roadImage.jpg')
```

```python
plt.imshow(img)
```

```python
# Get the S Threshold Value
hls = cv2.cvtColor(img, cv2.COLOR_RGB2HLS)

S = hls[:,:,2]
```

```python
thresh_S = (170,255)
S_threshold = np.zeros_like(S)
S_threshold[(S > thresh_S[0]) & (S <= thresh_S[1])] = 1
```

```python
plt.imshow(S_threshold, cmap='gray')
```

```python
# Get the Sobel_x Gradient Value
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
sobel_x = cv2.Sobel(gray, cv2.CV_64F, 1, 0)
```

```python
sobelx_abs = np.absolute(sobel_x)
```

```python
scaled_sobel = np.uint8(255*sobelx_abs/np.max(sobelx_abs))
```

```python
# Sobel Threshold
thresh_Sobel = (20,100)
Sob_threshold = np.zeros_like(scaled_sobel)
Sob_threshold[(scaled_sobel > thresh_Sobel[0]) & (scaled_sobel <= thresh_Sobel[1])] = 1
```

```python
plt.imshow(Sob_threshold, cmap='gray')
```

```python
# Combine Sobel Gradients and S Threshold Values to get a Single Output Image

# Stack each channel to view their individual contributions in green and blue respectively
# This returns a stack of the two binary images, whose components you can see as different colors
color_binary = np.dstack(( np.zeros_like(S_threshold), S_threshold, Sob_threshold)) * 255
```

```python
combined_binary = np.zeros_like(S_threshold)
combined_binary[(Sob_threshold == 1) | (S_threshold == 1)] = 1
```

```python
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(20,10))
ax1.set_title('Stacked thresholds')
ax1.imshow(color_binary)

ax2.set_title('Combined S channel and gradient thresholds')
ax2.imshow(combined_binary, cmap='gray')
```
