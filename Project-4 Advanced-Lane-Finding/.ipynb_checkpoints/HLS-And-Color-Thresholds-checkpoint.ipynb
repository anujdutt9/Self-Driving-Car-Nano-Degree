## HLS and Color Thresholds

```python
# Import Dependencies
import numpy as np
import cv2
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
%matplotlib inline
```

```python
# Read Image
img = mpimg.imread('./roadImage.jpg')
```

```python
plt.imshow(img)
```

```python
# Set the Threshold Value
thresh = (180, 255)
```

```python
# Convert to Grayscale
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
```

```python
plt.imshow(gray, cmap='gray')
```

```python
binary = np.zeros_like(gray)
binary[(gray > thresh[0]) & (gray <= thresh[1])] = 1
```

```python
plt.imshow(binary, cmap='gray')
```

## RGB Channels

```python
# Plotting R, G, B channels of the image
R = img[:,:,0]
G = img[:,:,1]
B = img[:,:,2]

fig, ax = plt.subplots(nrows=1, ncols=4, figsize=(20,30))
ax[0].set_title('Input Image')
ax[0].imshow(img)
ax[1].set_title('R Channel')
ax[1].imshow(R, cmap='gray')
ax[2].set_title('G Channel')
ax[2].imshow(G, cmap='gray')
ax[3].set_title('B Channel')
ax[3].imshow(B, cmap='gray')
```

```python
# Since, R Channel oes a good job of representing the Yellow Lane Line, we can apply threshold to it
thresh = (200, 255)
binary = np.zeros_like(R)
binary[(R > thresh[0]) & (R <= thresh[1])] = 1
```

```python
plt.imshow(binary, cmap='gray')
```

## HLS Channels

```python
hls = cv2.cvtColor(img, cv2.COLOR_RGB2HLS)

H = hls[:,:,0]
L = hls[:,:,1]
S = hls[:,:,2]
```

```python
fig, ax = plt.subplots(nrows=1, ncols=4, figsize=(20,30))
ax[0].set_title('Input Image')
ax[0].imshow(img)
ax[1].set_title('H Channel')
ax[1].imshow(H, cmap='gray')
ax[2].set_title('L Channel')
ax[2].imshow(L, cmap='gray')
ax[3].set_title('S Channel')
ax[3].imshow(S, cmap='gray')
```

```python
#The S channel picks up the lines well, so let's try applying a threshold there:
thresh = (90, 255)
binary = np.zeros_like(S)
binary[(S > thresh[0]) & (S <= thresh[1])] = 1
```

```python
plt.imshow(binary, cmap='gray')
```

```python
# In the H channel, the lane lines appear dark, so we could try a low threshold there.
thresh = (15, 100)
binary = np.zeros_like(H)
binary[(H > thresh[0]) & (H <= thresh[1])] = 1
```

```python
plt.imshow(binary, cmap='gray')
```
