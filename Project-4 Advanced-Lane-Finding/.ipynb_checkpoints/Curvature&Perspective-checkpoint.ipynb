```python
# Warp Function

def warp_image(img):
    img_size = (img.shape[1], img.shape[0])
    
    # Source Points
    src = np.float32([[1149,390],
                      [1618,382],
                      [1619,612],
                      [1151,612]])
    
    # Destination Recctangle Points
    dest = np.float32([[600,390],
                      [1060,390],
                      [1060,623],
                      [600,623]])
        
    # Compute Perspective Transform
    # Input Source & Dest Points
    # Returns Perspective Matrix
    M = cv2.getPerspectiveTransform(src, dest)
    
    # Inverse Perspective Transform to unWarp the Image
    # Minv = cv2.getPerspectiveTransform(dest, src)
    
    
    # Apply Transform "M" to Original Image to get the Destination Image
    # Inputs: image, Transform Matrix, image Output size, Interpolation Type
    # Interpolation: Filling in missing points in warped image
    warped_image = cv2.warpPerspective(img, M, img_size, flags=cv2.INTER_LINEAR)
    
    return warped_image
```

```python
# Test the Image Warp and Plot it
warped_img = warp_image(img)

fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(20,10))

ax[0].set_title('Source Image')
ax[0].imshow(img)
ax[1].set_title('Warped Image')
ax[1].imshow(warped_img)
```
