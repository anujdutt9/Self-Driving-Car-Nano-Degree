## Sobel Filter for Lane Line Detection

```python
import cv2
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.image import imread
%matplotlib inline
```

```python
# Read in the Road Image
# By default, "imread" reads the image as "BGR".
img = imread('./road.jpg')
```

```python
plt.imshow(img)
```

```python
# Convert Image tp Grayscale
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

plt.imshow(gray, cmap='gray')
```

```python
# Apply Sobel Filter in "x" direction
# For Sobel in x direction, use "1, 0" at end.
sobel_x = cv2.Sobel(gray, cv2.CV_64F, 1, 0)

plt.imshow(sobel_x, cmap='gray')
```

```python
# Apply Sobel Filter in "y" direction
# For Sobel in y direction, use "0, 1" at end.
sobel_y = cv2.Sobel(gray, cv2.CV_64F, 0, 1)

plt.imshow(sobel_y, cmap='gray')
```

```python
# Absolute Value of "x" Derivative
abs_sobelx = np.absolute(sobel_x)
```

```python
# Convert Absolute Image Value to 8-bit
scaled_sobel = np.uint8(255*abs_sobelx/np.max(abs_sobelx))
```

```python
# Create a binary threshold to select pixels based on gradient strength
thresh_min = 20
thresh_max = 100
sxbinary = np.zeros_like(scaled_sobel)
sxbinary[(scaled_sobel >= thresh_min) & (scaled_sobel <= thresh_max)] = 1
plt.imshow(sxbinary, cmap='gray')
```

So, in the above image, the pixel values are between 0 and 1 based on whether they lie in the threshold range of 20 to 100 or not. If they do, they are represented by a "1" else a "0".
The image above was converted to 8-bit representation i.e. 0-255 pixels.
