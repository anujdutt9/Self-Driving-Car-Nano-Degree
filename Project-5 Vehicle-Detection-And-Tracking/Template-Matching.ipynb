```python
# Import Dependencies
import cv2
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.image import imread
%matplotlib inline
```

```python
# Read Image
img = imread('./dataset/cutouts/bbox-example-image.jpg')
```

```python
plt.imshow(img)
```

```python
# Image Templates
templist = ['./dataset/cutouts/cutout1.jpg', './dataset/cutouts/cutout2.jpg', './dataset/cutouts/cutout3.jpg',
            './dataset/cutouts/cutout4.jpg', './dataset/cutouts/cutout5.jpg', './dataset/cutouts/cutout6.jpg']
```

```python
# Draw Boxes Function
def draw_boxes(img, bboxes, color=(0, 0, 255), thick=6):
    # Make a copy of the image
    imcopy = np.copy(img)
    # Iterate through the bounding boxes
    for bbox in bboxes:
        # Draw a rectangle given bbox coordinates
        cv2.rectangle(imcopy, bbox[0], bbox[1], color, thick)
    # Return the image copy with boxes drawn
    return imcopy
```

```python
# Function to take an image and a list of templates as inputs
# then searches the image and returns the a list of bounding boxes 
# for matched templates
def find_matches(img, template_list):
    # Make a copy of the image to draw on
    img2 = np.copy(img)
    # Define an empty list to take bbox coords
    bbox_list = []
    method = cv2.TM_CCOEFF
    # Iterate through template list
    for template in template_list:
        t = imread(template)
        res = cv2.matchTemplate(img2,t,method)
        min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(res)
        w, h = (t.shape[1], t.shape[0])
        if method in [cv2.TM_SQDIFF, cv2.TM_SQDIFF_NORMED]:
            top_left = min_loc
        else:
            top_left = max_loc
        bottom_right = (top_left[0] + w, top_left[1] + h)
        bbox_list.append((top_left, bottom_right))
    return bbox_list
```

```python
bboxes = find_matches(img, templist)
result = draw_boxes(img, bboxes)
plt.imshow(result)
```
