### Function to Resize the Image and Convert to Feature Vector

```python
def bin_spatial(img, color_space='RGB', size=(32,32)):
    img = imread(img)
    # Choose Color Space and Convert Image to That
    if color_space =='HSV':
        img = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)
    elif color_space == 'HLS':
        img = cv2.cvtColor(img, cv2.COLOR_RGB2HLS)
    elif color_space == 'Lab':
        img = cv2.cvtColor(img, cv2.COLOR_RGB2Lab)
    
    # Resize Image
    resized_img = cv2.resize(img, size)
    
    # Get Feature Vector
    feature_vec = resized_img.ravel()
    return feature_vec
```

```python
features = bin_spatial('./dataset/000275.png', 'HSV', size=(32,32))
```
